```python
from qiskit import QuantumCircuit
import matplotlib.pyplot as plt

###############################################################
# Toy Example
###############################################################

sigma0 = "1101"
sigma1 = "0110"

###############################################################
# Quantum Circuit
###############################################################

qc = QuantumCircuit(5, 5)

###############################################################
# Step 1: Prepare Superposition
###############################################################

qc.h(0)
qc.barrier(label="PKGen")

###############################################################
# Step 2: Encode |0>|σ0>
###############################################################

qc.x(0)

if sigma0[0] == "1":
    qc.cx(0,1)

if sigma0[1] == "1":
    qc.cx(0,2)

if sigma0[2] == "1":
    qc.cx(0,3)

if sigma0[3] == "1":
    qc.cx(0,4)

qc.x(0)

###############################################################
# Step 3: Encode |1>|σ1>
###############################################################

if sigma1[0] == "1":
    qc.cx(0,1)

if sigma1[1] == "1":
    qc.cx(0,2)

if sigma1[2] == "1":
    qc.cx(0,3)

if sigma1[3] == "1":
    qc.cx(0,4)

qc.barrier(label="Verification")

###############################################################
# Step 4: Phase Encoding (b = 1)
###############################################################

qc.z(0)

qc.barrier(label="Phase")

###############################################################
# Step 5: Hadamard Measurement
###############################################################

for i in range(5):
    qc.h(i)

qc.barrier(label="Hadamard")

###############################################################
# Step 6: Measurement
###############################################################

qc.measure(range(5), range(5))

###############################################################
# Draw Colorful Circuit
###############################################################

fig = qc.draw(
    output="mpl",
    style="iqp",
    fold=-1,
    idle_wires=False
)

plt.show()
```

```python
fig = qc.draw(
    output="mpl",
    style="iqp",
    fold=-1,
    idle_wires=False
)

fig.savefig(
    "QPKBE_Circuit.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
